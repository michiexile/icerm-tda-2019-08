```python
%pylab inline
import numpy as np
```

```python
eg = 'Example1_d=0.5'
points = np.loadtxt(eg+'.txt')
theta = [2*pi-arccos(x/norm([x,y], ord=2)) if arcsin(y/norm([x,y], ord=2))<0 else arccos(x/norm([x,y], ord=2)) for x,y in points]
for l in ['0','0.5','1']:
    gcc = np.loadtxt(eg+'_CircularCoordinates_'+l+'_0.txt')
    figure(figsize=(5,5),dpi=100)
    scatter(theta,gcc,c='b',s=10)
    savefig(eg+'-'+l+'.pdf',format='pdf')
```
