```python
%pylab inline
import numpy as np
```

```python
plt.figure(figsize=(15,15))

for sonar in [3,4,5]:
    for lamb in ['0','0.5','1']:
        data = np.loadtxt('collection_{}_tshort_CircularCoordinates_'.format(sonar)+lamb+'_0.txt')
        
        #
        i = ['0','0.5','1'].index(lamb)*3+sonar-2
        plt.subplot(3,3,i)
        x = cos(data*2*pi)
        y = sin(data*2*pi)
        scatter(x,y,c='b')
        plt.xlim(-1.1,1.1)
        plt.ylim(-1.1,1.1)
        gca().set_aspect(1)
        
        if lamb == '1':
            plt.xlabel('collection {}'.format(sonar))
        if sonar == 3:
            plt.ylabel('lambda='+lamb)
plt.savefig('Fig4.1.pdf',format='pdf')
```
