```python
import numpy as np
import PCAtool
X=np.random.rand(300,2)
print(X)
#Create a random matrix as fake data.
X_new=PCAtool.pca(X,K=2)
print(X_new)
```

```python
import matplotlib.pyplot as plt
X_cen=X-np.mean(X,axis=0)
plt.scatter(X_cen[:,0], X_cen[:,1])
plt.scatter(X_new[:,0], X_new[:,1])
plt.show()
```
